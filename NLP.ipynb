```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('train.txt' , sep = ';' , header=None , names=['text' , 'emotion'])
```

```python
# Get unique emotion values
unique_emotions = df['emotion'].unique()

# Create empty dictionary
emotion_numbers = {}

# Counter variable
i = 0

# Assign number to each emotion
for emo in unique_emotions:
    emotion_numbers[emo] = i
    i += 1

# Map emotions to numbers
df['emotion'] = df['emotion'].map(emotion_numbers)

# Display mapping
print(emotion_numbers)

# Preview updated dataframe
df.head()
```

```python
# Convert text column to lowercase
df['text'] = df['text'].apply(lambda x: x.lower())

#
```

```python
# 2. Remove all punctuation using str.translate
# We create a translation table mapping punctuation characters to None (deleting them)
import string

def remove_punc(txt):
    return txt.translate(str.maketrans('', '', string.punctuation))

```

```python
df['text'] = df['text'].apply(remove_punc)
```

```python
def remove_numbers(txt):
    new = ""
    for i in txt:
        if not i.isdigit():
            new = new + i
    return new
```

```python
df['text'] = df['text'].apply(remove_numbers)
```

```python
def remove_emojis(txt):
    new = ""
    for i in txt:
        if i.isascii():
            new += i
    return new
df['text'] = df['text'].apply(remove_numbers)
```

```python
import nltk

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('punkt_tab') # Added to resolve LookupError

stop_words = set(stopwords.words('english'))
```

```python
df.loc[1]['text']
```

```python
def remove_stopwords(txt):
    words = word_tokenize(txt)
    cleaned = []
    for word in words:
        if word not in stop_words:
            cleaned.append(word)
    return ' '.join(cleaned) # Corrected return statement
```

```python
df['text'] = df['text'].apply(remove_stopwords)
```
