# **Bag of Words(BoW)**


Bag of Words is a Natural Language Processing (NLP) technique used to convert text data into numerical form so machine learning models can understand it.

How Bag of Words Works

1. Collect all unique words from the text dataset.

2. Create a vocabulary of these words.

3. Count how many times each word appears in a sentence or document.

4. Represent each sentence as a vector of word frequencies.

Example

Sentence 1: "I love AI"
Sentence 2: "AI is powerful"

Vocabulary:
["I", "love", "AI", "is", "powerful"]

Vector Representation:
Sentence 1 → [1, 1, 1, 0, 0]
Sentence 2 → [0, 0, 1, 1, 1]

Why Bag of Words is Used

1. Converts text into numerical format.

2. Simple and easy to implement.

3. Works well for text classification tasks.

4. Commonly used in sentiment analysis and spam detection.

Advantages

1. Easy to understand.
2. Fast to implement.
3. Works well on small datasets.

Disadvantages

1. Ignores word order.
2. Does not understand context.
3. Creates sparse matrices for large vocabularies.
4. Cannot capture semantic meaning of words.

```python
from sklearn.feature_extraction.text import CountVectorizer

documents = [
    "I love pizza",
    "Pizza is the best",
    "I love pasta",
    "Pasta is great"
]

vectorizer = CountVectorizer()

X = vectorizer.fit_transform(documents)

print("Vocabulary:", vectorizer.get_feature_names_out())
print("\nBoW Matrix:\n", X.toarray())
```
